```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch 
import seaborn as sns
```

```python
# Read in the data
df=pd.read_csv(r"C:\Users\preci\Downloads\messibetis.csv")
```

```python
df.head()
```

```python
df['x']=df['x']*1.2
df['y']=df['y']*0.8
df['endX']=df['endX']*1.2
df['endY']=df['endY']*0.8
```

```python
df.head()
```

```python
# Set up the pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')
plt.gca().invert_yaxis()
# Plot the two points
# plt.plot([x1, x2], [y1, y2], marker='o', linestyle='-')
plt.title('Messi Passes VS Real Betis',color='white',size=30)
legend_list_sf=[]
legend_list_usf=[]
for i in range(len(df['x'])):
    if df['outcome'][i]=='Successful':
        legend_list_sf.append('Successful')
        if len(legend_list_sf)>1:
            plt.plot((df['x'][i],df['endX'][i]),(df['y'][i],df['endY'][i]),color="green")
            plt.scatter(df['x'][i],df['y'][i],color='green')
            
        else:
            plt.plot((df['x'][i],df['endX'][i]),(df['y'][i],df['endY'][i]),color="green",label='Successful Pass')
            plt.scatter(df['x'][i],df['y'][i],color='green',label='Start Point')
            
    if df['outcome'][i]=='Unsuccessful':
        legend_list_usf.append('Unsuccessful')
        if len(legend_list_usf)>1:
            plt.plot((df['x'][i],df['endX'][i]),(df['y'][i],df['endY'][i]),color="red")
            plt.scatter(df['x'][i],df['y'][i],color='red')
        else:
            plt.plot((df['x'][i],df['endX'][i]),(df['y'][i],df['endY'][i]),color="red",label='Unsuccessful Pass')
            plt.scatter(df['x'][i],df['y'][i],color='red',label='Start Point')

plt.legend()
plt.show()
        
```
